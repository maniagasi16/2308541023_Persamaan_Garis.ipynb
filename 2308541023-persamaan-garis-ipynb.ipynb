```python
import numpy as np
import matplotlib.pyplot as plt

# Persamaan garis melewati titik (a, b) dan (c, d)
a, b = 1, 2
c, d = 4, 5

# Menghitung gradien m
m = (d - b) / (c - a)
print(f"Gradien garis adalah {m}")

# Menentukan persamaan garis
# y - b = m(x - a)
# y = m(x - a) + b
# y = mx - ma + b
r = -m * a + b
print(f"Persamaan garis: y = {m}x + {r}")

# Menggambar garis
x = np.linspace(0, 5, 100)
y = m * x + r

plt.figure(figsize=(8, 6))
plt.plot(x, y, label=f'y = {m:.2f}x + {r:.2f}', color='blue')
plt.scatter([a, c], [b, d], color='red', zorder=5)  # Titik-titik (a, b) dan (c, d)
plt.text(a, b, f'({a}, {c})', fontsize=12, ha='right')
plt.text(c, d, f'({c}, {d})', fontsize=12, ha='left')

# Mengatur sumbu dan judul
plt.xlabel('x')
plt.ylabel('y')
plt.title('Grafik Persamaan Garis')
plt.legend()
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Persamaan garis pertama melewati titik (a,b) dan (c, d) 
a, b = 2, 3
c, d = 0, 1

# Persamaan garis ke dua melewati titik (e,f) dan (g,h)
e, f = 4, 5
g, h = 2, 0

# Menghitung gradien untuk garis pertama
m1 = (d - b) / (c - a)
c1 = b - m1 * a
print("Gradien garis pertama adalah {}".format(m1))
print("Persamaan garis pertama adalah y = {}x + {}".format(m1, c1))

# Menghitung gradien untuk garis kedua
m2 = (h - f) / (g - e)
c2 = f - m2 * e
print("Gradien garis kedua adalah {}".format(m2))
print("Persamaan garis kedua adalah y = {}x + {}".format(m2, c2))

if m1==m2:
    print("Gradiennya tidak sama, jadi ada kemungkinan garis-garis tersebut berpotongan.")
else:
    # Titik potong kedua garis pada (tp_x,tp_y)
    tp_x = (m1*a-b-m2*e+f)/(m1-m2)
    tp_y = m2*(tp_x-e)+f

    print("Titik potong untuk adalah ({},{})".format(tp_x,tp_y))

    x=np.arange(tp_x-5,tp_x+5,0.01)
    plt.figure(figsize=(4,4),dpi=500)
    
    plt.plot(x,m1*(x-a)+b,color='red',label='y={}x+{}'.format(m1,-m1*a+b))
    plt.plot(x,m2*(x-e)+f,color='blue',label='y={}x+{}'.format(m2,-m2*e+f))
    plt.plot(tp_x,tp_y,color='green',marker='o',markersize=4)
    plt.legend(loc='best')
    plt.show()
```
